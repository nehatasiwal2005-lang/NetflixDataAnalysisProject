```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titles.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="type", data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df["release_year"], bins=30)
plt.title("Content Release Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
top_genres = df["genres"].value_counts().head(10)

plt.figure(figsize=(8,5))
top_genres.plot(kind="bar")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["imdb_score"], bins=20)
plt.title("IMDb Score Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["tmdb_score"], bins=20)
plt.title("TMDB Score Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["runtime"], bins=30)
plt.title("Runtime Distribution")
plt.show()
```

```python
top_countries = df["production_countries"].value_counts().head(10)

plt.figure(figsize=(8,5))
top_countries.plot(kind="bar")
plt.title("Top Production Countries")
plt.xlabel("Country")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="imdb_score", y="tmdb_score", data=df)
plt.title("IMDb Score vs TMDB Score")
plt.show()
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```

```python
import pandas as pd
```

```python
df = pd.read_csv("titles.csv")
```

```python
df = df.drop_duplicates()
```

```python
df["imdb_score"] = df["imdb_score"].fillna(df["imdb_score"].mean())
df["tmdb_score"] = df["tmdb_score"].fillna(df["tmdb_score"].mean())
df["tmdb_popularity"] = df["tmdb_popularity"].fillna(df["tmdb_popularity"].mean())
```

```python
df.to_csv("netflix_cleaned.csv", index=False)
```
